```python
# Imports
from wordcloud import WordCloud
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import pandas as pd
import nltk
```

```python
# Lendo o arquivo com as legendas de cada post no Instagram do FEA.dev
df = pd.read_excel('post_caption_dataset.xlsx')
```

```python
# Baixando stopwords do pacote NLTK
nltk.download('stopwords')

# Usando stopwords em português
stopwords = set(stopwords.words('portuguese'))

# Adicionando a palavra "sobre" ao conjunto de stopwords
stopwords.add("sobre")

# Combinando todas as legendas de publicações em um único texto
text = ' '.join(df['Post caption']).lower()

# Criando um objeto WordCloud
wordcloud = WordCloud(stopwords=stopwords, # conjunto com palavras que serão ignoradas
                      background_color='white', # cor de fundo da imagem
                      colormap='plasma', # paleta de cores das palavras
                      width=800, # largura da imagem
                      height=400 # altura da imagem
                      ).generate(text) # método que gera a nuvem de palavras

# Visualizando a nuvem de palavras
plt.figure(figsize=(10, 5)) # definindo tamanho da figura
plt.imshow(wordcloud, interpolation='bilinear') # suavizando a imagem
plt.axis("off") # removendo eixos
plt.show()
```
